The uniform distribution of position of neutrons

The cross-sectional ring area at radius $r$ scales as $2\pi rdr$ so for the uniform distribution, more neutrons are in the shell with larger $r$. For an extreme case $P_{\text{abs}}\approx 0.8$, most neutrons get absorbed in their first random walk, thus the final absorption positions increase roughly linearly with $r$.
